# Expfit notes 5: Multiple exponential heuristics

To fit more than 1 exponential, we use a few more heuristics.

## Fitd2: double decaying, same sign multiplier

1. An initial estimate for a single exponential is made on unit square transformed data. The 3 parameters are detransformed to $(a_0, b_0, c_0)$, and the remaining fit is on untransformed data.
2. Checks are run: The initial $c$ should be negative.
3. The area under the curve $A0$ of the _data minus a0_ is integrated.
4. $p_0[0] = a_0$, $p_0[2] = c_0 / \sqrt{2}$, $p_0[4] = c_0 \sqrt{2}$.
5. The area $A$ under $b_0 e^{p_0[2] t} + b_0 e^{p_0[2] t}$ is calculated directly
6. $p_0[1] = p_0[3] = b_0 A0 / A$
7. The fit is run to obtain $p$.
   
If the optimiser reports an error, or $p[4] / p[2] \leq 1.1$, the initial estimates for $p_0[2]$ and $p_0[4]$ are decreased/increased by a factor $\sqrt{2}$ and steps 5-7 are repeated (this is rare).
