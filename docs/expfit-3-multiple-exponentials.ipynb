# Expfit notes 3: Multiple exponentials

$E_m$ is the error for $m$ exponentials

As before, we'll use a shorthand 
\begin{align}
e_ij = e^{b_j x_i}
\end{align}
But for clarity, we'll omit the $i$ indices:
\begin{align}
e_j = e^{b_j x}
\end{align}

## Two exponentials

\begin{align}
E_2 &= \frac{1}{n}\sum^n \left( y - a - b_1 e_1 - b_2 e_2 \right)^2 &&  \\
    &= \frac{1}{n}\sum^n y^2 + a^2 - 2ay + 2(a - y)(b_1e_1 + b_2e_2) + b_1^2e_1^2 + b_2^2e_2^2 + 2 b_1b_2e_1e_2
\end{align}

$z \in {1, 2}$

### Jacobian $J_2$

\begin{align}
\frac{\partial E_2}{\partial a} &= \frac{2}{n}\sum^n a - y + b_1e_1 + b_2e_2
\end{align}

\begin{align}
\frac{\partial E_2}{\partial b_z} &= \frac{2}{n}\sum ^n\left( a - y + b_1 e_1 + b_2 e_2 \right) e_z
\end{align}

\begin{align}
\frac{\partial E_2}{\partial c_z} &= \frac{2b_z}{n}\sum^n \left( a - y + b_1 e_1 + b_2 e_2 \right) xe_z
\end{align}

### Hessian $H_2$

\begin{align}
\frac{\partial^2 E_2}{\partial a^2} &= 2 \\
\frac{\partial^2 E_2}{\partial b_z\,\partial a} &= \frac{2}{n}\sum^n e_z \\
\frac{\partial^2 E_2}{\partial c_z\,\partial a} &= \frac{2b_z}{n}\sum^n xe_z
\end{align}

\begin{align}
\frac{\partial^2 E_2}{\partial a\,\partial b_z} &= \frac{2}{n}\sum ^n\ e_z \\
\frac{\partial^2 E_2}{\partial b_z^2} &= \frac{2}{n}\sum ^n e_z^2 \\
\frac{\partial^2 E_2}{\partial c_z \partial b_z} &= \frac{2}{n}\sum ^n\left( a - y + b_1 e_1 + b_2 e_2 \right) x e_z + x b_z e_z^2
\end{align}

\begin{align}
\frac{\partial^2 E_2}{\partial a \partial c_z} &= \frac{2b_z}{n}\sum^n xe_z \\
\frac{\partial^2 E_2}{\partial b_z \partial c_z} &= \frac{2}{n}\sum^n \left( a - y + b_1 e_1 + b_2 e_2 \right) xe_z + xb_ze_z^2 \\
\frac{\partial^2 E_2}{\partial c_z^2} &= \frac{2b_z}{n}\sum^n \left( a - y + b_1 e_1 + b_2 e_2 \right) x^2e_z + x^2b_z^2e_z^2 \\
\end{align}

## General case

\begin{align}
E_m &= \frac{1}{n}\sum^n \left( y - a - \sum_{j=1}^m b_j e_j \right)^2 &&  \\
    &= \frac{1}{n}\sum^n \left( y^2 + a^2 - 2ay + 2\sum_{j=1}^m (a - y)b_je_j + \sum_{j=1}^m b_j^2e_j^2 + 2\sum_{j=1}^m\sum_{k=j+1}^m b_jb_ke_je_k \right)
\end{align}

### Jacobian

\begin{align}
\frac{\partial E_m}{\partial a} &= \frac{2}{n}\sum^n \left( a - y + \sum_{j=1}^m b_je_j \right)
\end{align}

\begin{align}
\frac{\partial E_m}{\partial b_z} &= \frac{2}{n}\sum ^n\left( a - y + \sum_{j=1}^m b_j e_j \right) e_z
\end{align}

\begin{align}
\frac{\partial E_m}{\partial c_z} &= \frac{2b_z}{n}\sum^n \left( a - y + \sum_{j=1}^m b_j e_j \right) xe_z
\end{align}

### Hessian

\begin{align}
\frac{\partial^2 E_2}{\partial a^2} &= 2 \\
\frac{\partial^2 E_2}{\partial b_z\,\partial a} &= \frac{2}{n}\sum^n e_z \\
\frac{\partial^2 E_2}{\partial c_z\,\partial a} &= \frac{2b_z}{n}\sum^n xe_z
\end{align}

\begin{align}
\frac{\partial^2 E_2}{\partial a\,\partial b_z} &= \frac{2}{n}\sum ^n\ e_z \\
\frac{\partial^2 E_2}{\partial b_z^2} &= \frac{2}{n}\sum ^n e_z^2 \\
\frac{\partial^2 E_2}{\partial c_z \partial b_z} &= \frac{2}{n}\sum ^n\left( a - y + \sum_{j=1}^m b_j e_j \right) x e_z + x b_z e_z^2
\end{align}

\begin{align}
\frac{\partial^2 E_2}{\partial a \partial c_z} &= \frac{2b_z}{n}\sum^n xe_z \\
\frac{\partial^2 E_2}{\partial b_z \partial c_z} &= \frac{2}{n}\sum^n \left( a - y + \sum_{j=1}^m b_j e_j \right) xe_z + xb_ze_z^2 \\
\frac{\partial^2 E_2}{\partial c_z^2} &= \frac{2b_z}{n}\sum^n \left( a - y + \sum_{j=1}^m b_j e_j \right) x^2e_z + x^2b_z^2e_z^2 \\
\end{align}
