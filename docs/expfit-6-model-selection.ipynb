# Expfit notes 6: Model selection

## Nested models: likelihood ratio test

Accept a model with $m + 1$ exponential terms over one with $m$ terms if

\begin{align}
\lambda = -2 \left( \ell(\hat\theta_m) - \ell(\hat\theta_{m + 1}) \right) \geq \chi^2_{1,1-\alpha}
\end{align}

Should be chi-square distributed with 1 dof

\begin{align}
\ell(\theta) = -\frac{n}{2}\log(2\pi) - n\log(\sigma) - \frac{n}{2\sigma^2} E(\theta)
\end{align}

On same data, using same $n$ and same (estimate of) $\sigma$

\begin{align}
\lambda = \frac{n}{\sigma^2} \left( E(\hat\theta_{m}) - E(\hat\theta_{m+1}) \right)
\end{align}

Using
\begin{align}
E(\hat\theta_{m}) - E(\hat\theta_{m+1}) < \frac{\sigma^2}{n} \chi^2_{1,1-\alpha}
\end{align}

## Non-nested models

Not as good as likelihood ratio, but can use the [Akaike information criterion](https://en.wikipedia.org/wiki/Akaike_information_criterion):
\begin{align}
\mathcal{A} 
= 2m - 2 \ell(\theta) 
= 2m + n\log(2\pi) + 2n\log(\sigma) + \frac{n}{\sigma^2} \hat{E}
\end{align}
where $m$ is the number of parameters in the model (1 for the offset, and then 2 per exponential).

A lower AIC indicates a better model.
So we select the model with the lowest 
\begin{align}
\mathcal{A}^* = 2m + \frac{n}{\sigma^2} \hat{E}_m
\end{align}
where $\hat{E}_m$ is the best solution found for the $m$-parameter model.

Note that we have to use the same $\sigma^2$ approximation for both models.

### Line versus exponential

Assume the exponential is a better fit: $E_3 < E_2$, then use $E_3 \approx \sigma^2$

Accept line if
\begin{align}
2 \cdot 2 + \frac{n}{\sigma^2} E_2 &< 2 \cdot 3 + \frac{n}{\sigma^2} E_3 \\
n \frac{E_2}{E_3} &< 2 + n \\
E_2 &< \frac{2 + n}{n} E_3
\end{align}

For large $n$ this quickly becomes very small, e.g. $n = 1000 \rightarrow 1.002$
