# Expfit notes 6: Model selection

Accept a model with $m + 1$ exponential terms over one with $m$ terms if

\begin{align}
\lambda = -2 \left( \ell(\hat\theta_m) - \ell(\hat\theta_{m + 1}) \right) \geq \chi^2_{1,1-\alpha}
\end{align}

Should be chi-square distributed with 1 dof

\begin{align}
\ell(\theta) = -\frac{n}{2}\log(2\pi) - n\log(\sigma) - \frac{n}{2\sigma^2} E(\theta)
\end{align}

On same data, using same $n$ and same (estimate of) $\sigma$

\begin{align}
\lambda = \frac{n}{\sigma^2} \left( E(\hat\theta_{m}) - E(\hat\theta_{m+1}) \right)
\end{align}

Using
\begin{align}
E(\hat\theta_{m}) - E(\hat\theta_{m+1}) < \frac{\sigma^2}{n} \chi^2_{1,1-\alpha}
\end{align}
